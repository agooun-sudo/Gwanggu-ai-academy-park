```python
import pandas as pd
import numpy as np

series = pd.Series({'사과':1200,'딸기':3500,'키위':2100})
print(series)

#단일값 접근
print(series['사과'])
print(series['딸기'])
# 펜시인덱싱 / 펜시가 시리즈 통으로가져옴 [] 접근지정자
print(series[ ['딸기','사과'] ])
#print(series[ ['키위'] ])

# 과일주에 2000 원 이상 과일을 시리즈로 뽑아내시오
# 조건 필터링을 통한 접근 (불린 인덱싱) => 시리즈도 넘파이 처럼 불린인덱싱이 지원됨
print(series[series >= 2000])
```

```python
# 행번호 인덱싱
print(series[1:3])
print(series[:])



#인덱스기반 인덱싱
print(series['사과':'키위'])
print(series['사과':'딸기'])

#iloc 함수 활용 i = integer

print(series.iloc[0:3])
```

```python
#iloc 함수 활용 i = integer
print(series.iloc[0:3])

print(series.iloc[1:3])

#특정 위치에 있는 값 가져오기
print(series.iloc[0])

# 시리즈 추가
# 없으면 추가 , 있으면 업데이트
# 인덱스 이름으로
series.loc['바나나'] = 5000
print(series)

series.loc['참외'] = 6000
print(series)

# 업데이트
series.loc['참외'] = 7000
print(series)
series.iloc[-1] = 9000
print(series)

#삭제

# series.drop('사과')
# print(series)
```

```python
#삭제
#inplace 원본수정
# series.drop('사과',inplace = True)
# print(series)
# series.drop(['딸기','바나나'],inplace = True)
# print(series)
# series_2 = series.drop('사과')
# print(series_2)
```

```python
series.drop(['딸기','바나나'],inplace = True)
print(series)
```
