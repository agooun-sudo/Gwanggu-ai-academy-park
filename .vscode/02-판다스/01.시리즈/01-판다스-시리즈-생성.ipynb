# excel 문서의 시트와 유사
# 행 과 열로 구성
# 각 열별로 데이타 타입을 가짐
# 시리즈 + 컬럼


# 모든 길은 로마로 통하듯이, 모든 데이타는 판다스로 통한다.
# 판다스 = 시리즈 객체 + 데이타 프레임 객체


<!-- 차원 배열 구조 - Series : 1차원 배열
행번호 + 인덱스 + 데이타 3개의 변수를 기본적으로 가지고 있음
인덱스(index) 사용 가능 - 인덱싱/슬라이싱
데이터 타입을 가짐 (dtype)
모든 길은 로마로 통하듯이, 모든 데이타는 판다스로 통한다.
판다스 = 시리즈 객체 + 데이타프레임 객체 -->

```python
import pandas as pd #pip install pandas
import numpy as np
# 1차원 배열 시리즈 객체 생성
arr = np.arange(100,105)
print(arr)
series = pd.Series(arr)
print(series)
print(dir(series))
```

```python

arr = np.arange(100,105)
print(arr)

series = pd.Series(arr,dtype='float')
print(type(series))
print(series)
print(dir(series))
```

```python

series = pd.Series((100,200,300,400),dtype='float')
print(type(series))
print(series)
print(dir(series))
```

```python

series = pd.Series(["사과","딸기"])
print(type(series))
print(series)
print(dir(series))
```

```python
series = pd.Series(["사과",91,3.14])
print(type(series))
print(series)
print(dir(series))
```

```python
data = {'a': 100,'b':200}

series = pd.Series(data)
print(series)
```
