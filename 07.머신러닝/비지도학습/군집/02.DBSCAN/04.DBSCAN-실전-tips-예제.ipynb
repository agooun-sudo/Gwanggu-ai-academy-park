```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset("tips")

# 사용할 컬럼
df = df[["total_bill", "tip", "size"]]

print(df.head())


```

```python
from sklearn.preprocessing import StandardScaler
x = df[["total_bill", "tip", "size"]]
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
from sklearn.cluster import DBSCAN

model = DBSCAN(eps=0.5,min_samples=5)
df["cluster"] = model.fit_predict(x_scaled)

```

```python
import matplotlib.pyplot as plt

plt.scatter(df["total_bill"], df["tip"], c=df["cluster"])
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("DBSCAN Clustering (Outlier Detection)")
plt.show()
```

```python
print(df["cluster"].value_counts())
```

```python
print(df.groupby("cluster").mean(numeric_only=True))
# 분석
# 0 => 소규모 / 저가 소비 고객
# 1 => 중간규모 / 일반 소비 고객
# 2 => 대규모 / 고가 소비 고객
# -1 => 이상치 고객 => 이것만 모아서 따로 분석 가능
```

```python
# 1. DBSCAN 을 적용하여 아래와 같이 출력이 나오도록 하시오.
# 전처리시 정규화 적용 할것)

# import numpy as np
# import pandas as pd

# np.random.seed(42)




import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN

# --------------------------
# 데이터 생성
# --------------------------
np.random.seed(42)

df = pd.DataFrame({
    "x": np.random.randint(0, 100, 30),
    "y": np.random.randint(0, 100, 30)
})

# --------------------------
# 정규화
# --------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

# --------------------------
# DBSCAN 적용
# --------------------------
model = DBSCAN(eps=0.8, min_samples=3)
model.fit(X_scaled)

# --------------------------
# 군집 결과 저장
# --------------------------
df["cluster"] = model.labels_

# 결과 출력
print(df)

# 군집별 개수
print(df["cluster"].value_counts().sort_index())

# 군집별 평균
print(df.groupby("cluster").mean())
```
