```python
# 입력 변수들끼리
# 모델이 어떤 변수가 진짜 영향을 주는지 구분하기 어려운 상태

import pandas as pd

df = pd.DataFrame({
    "study":[1,2,3,4,5,6,7,8],
    "sleep":[8,7,7,6,6,5,5,4],
    "homework":[2,4,6,8,10,12,14,16],
    "score":[52,58,65,70,76,82,90,95] # 종속 
})

print(df)
```

```python
# 2. 회귀모델 만들기
import statsmodels.api as sm

x = df[["study","sleep","homework"]]
y = df["score"]

x = sm.add_constant(x)
model = sm.OLS(y, x)
result = model.fit()
print(result.summary())
```
