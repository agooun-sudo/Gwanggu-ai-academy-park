```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Input
import numpy as np
import matplotlib.pyplot as plt
```

```python

x = np.array([[2,0],[4,4],[6,2],[8,3]])
y = np.array([81,93,91,97])
```

```python
model = Sequential([
    Input(shape=(2,)), # мһ…л Ҙ нҠ№м„ұ(feature)мқҙ 2к°ң /м»¬лҹј
    Dense(1,activation="linear")
])

model.compile(optimizer="sgd",loss="mse")


model.fit(x,y, epochs=1000)
```

```python
weights, bias = model.layers[0].get_weights()

print(weights)
print(bias)
```

```python
model.summary()
```

```python
pred = model.predict(np.array([[10,5],[11,6]]))
print(pred)
```

```python

# кІҪмӮ¬н•ҳк°•лІ• м§Ғм ‘ кө¬нҳ„

# мҳӨм°Ё кі„мӮ°
# вҶ“

# кё°мҡёкё° кі„мӮ°
# вҶ“

# к°ҖмӨ‘м№ҳ мҲҳм •
# вҶ“

# л°ҳліө

w = 10
lr = 0.1

for epoch in range(20):

    loss = (w-3)**2 
    
    grad = 2 * (w-3) # мң„мқҳ лҜёл¶„
    
    #мғҲ к°ҖмӨ‘м№ҳ = кё°мЎҙ к°ҖмӨ‘м№ҳ - н•ҷмҠөлҘ Г—кё°мҡёкё°
    w = w - lr * grad 

    print(epoch,round(w,3),round(loss,3))
```

```python
# л¬ём ң  мһҗлҸҷм°Ё мҶҚлҸ„мҷҖ м—°л№„

# лҚ°мқҙн„°

x = np.array([
    40,
    60,
    80,
    100,
    120
])

y = np.array([
    25,
    22,
    19,
    16,
    13
])
# н•ҙм•ј н•  мқј
# лӘЁлҚё мғқм„ұ
# н•ҷмҠө мҲҳн–ү
# мҶҚлҸ„ 90kmмқј л•Ң м—°л№„ мҳҲмёЎ

# мҳӣлӮ  мҶҢмҠӨмҪ”л“ң
# м¶ңл Ҙ к°’, мһ…л Ҙ ліҖмҲҳ, л¶„м„қ л°©лІ•м—җ л§һкІҢлҒ” лӘЁлҚёмқ„ м„Өм •н•©лӢҲлӢӨ. 
# model.add(Dense(1, input_dim=2, activation='linear'))
# model.add(Dense(1, input_shape=(2,), activation='linear'))
# model.add(Dense(1, Input(shape=(2,)), activation='linear'))

model = Sequential([
    Input(shape=(1,)), # мһ…л Ҙ нҠ№м„ұ(feature)мқҙ 1к°ң /м»¬лҹј
    Dense(1,activation="linear")
])

model.compile(optimizer="adam",loss="mse")

model.fit(x,y, epochs=2000,verbose=1)

```

```python
pred = model.predict(np.array([90]))
print(pred)
```

```python
# к°ҖмӨ‘м№ҳ л°”мқҙм–ҙмҠӨ нҷ•мқё
weights, bias = model.layers[0].get_weights()

print(weights, bias)

pred = model.predict(np.array([90]))
print(pred)
```

```python


# лӢӨн•ӯ л¬ём ң
# л¬ём ң  кҙ‘кі л№„ + мҳҒм—…мӮ¬мӣҗмҲҳ вҶ’ л§Өм¶ң

import numpy as np

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense

# ==========================
# лҚ°мқҙн„°
# кҙ‘кі л№„, мҳҒм—…мӮ¬мӣҗмҲҳ
# ==========================

x = np.array([
    [10, 2],
    [20, 3],
    [30, 4],
    [40, 5]
])

# л§Өм¶ң
y = np.array([
    100,
    180,
    260,
    340
])

# мҳҲмёЎ
# кҙ‘кі л№„=35
# мҳҒм—…мӮ¬мӣҗмҲҳ=4
```

```python
model = Sequential([
    Input(shape=(2,)), # мһ…л Ҙ нҠ№м„ұ(feature)мқҙ 1к°ң /м»¬лҹј
    Dense(2,activation="linear"),
    Dense(1,activation="linear")
])

model.compile(optimizer="adam",loss="mse")

model.fit(x,y, epochs=1000,verbose=1)

```

```python
weights, bias = model.layers[0].get_weights()

print(weights)
print(bias)
```

```python
# мҳҲмёЎ
pred = model.predict(np.array([[35,4]]))
print(pred)
```
