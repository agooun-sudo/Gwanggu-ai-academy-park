```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
import numpy as np


# нӮӨ + лӘёл¬ҙкІҢ вҶ’ л№„л§Ң м—¬л¶Җ

X = np.array([
    [170,60],
    [175,65],
    [180,70],
    [170,90],
    [175,95],
    [180,100]
])

y = np.array([
    0,
    0,
    0,
    1,
    1,
    1
])
```

```python
#лӘЁлҚё мғқм„ұ

model = Sequential([
    Input(shape=(2,)),
    Dense(1,activation="sigmoid")
])

model.compile(
            optimizer="adam",
            loss="binary_crossentropy",
            metrics=["accuracy"]
)

#н•ҷмҠө
model.fit(X,y,epochs=1000,verbose=1)
```

```python
pred = model.predict(np.array([[172,88]]))

print(pred)

if pred[0][0] >= 0.5:
    print("л№„л§Ң")
else:
    print("м •мғҒ")
```

```python
model.summary()
```
